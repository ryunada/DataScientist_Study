### 2-2. DataFrame 생성 :: pd.DataFrame( )

```python

    keys,               # label or list of labels
    drop = True,        # True : 새 인덱스로 사용할 열 삭제 : False : 새 인덱스로 사용할 열 보존
    inplace = False,    # True : 원본 대체 | False : 원본 유지

```
- 기존 열을 사용하여 DataFrame 인덱스를 설정
- Option
